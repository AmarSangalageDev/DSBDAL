```python
df['Salaryroup']=pd.cut(df['Salary'],bins=[25000,50000,75000,100000,125000],labels=['25000-50000','50000-75000','75000-100000','100000-125000'])
```

```python
grouped=df.groupby('Salaryroup',observed=True)['YearsExperience']
```

```python
summary=grouped.agg(['mean','median','min','max','std'])
```

```python
summary
```
